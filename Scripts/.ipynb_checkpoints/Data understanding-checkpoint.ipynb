## Understanding the magnitude of our problem <a name="Understanding"></a>

In this project we will use `machine learning` to develop a model that can `warn the driver` for the posibility of them getting into a `car accident`, and its `possible severity`, given the `weather and road conditions`. This could help the driver pay more attention when driving in areas where the probability of an accident is higher, or even change the route to a safer one. 

Car accidents are still the `leading cause of death for Americans under 30`[[1](https://www.cdc.gov/winnablebattles/report/motor.html)], and they also take `billions of dolars off the US economy`[[2](https://crashstats.nhtsa.dot.gov/Api/Public/ViewPublication/812013)]. This is why it is of vital importance to develp a model that can prevent these accidents from happening in the first place, and, hopefully, save as many lives as possible.



## Getting our data <a name="Getting_data"></a>

The City of Seattle has an "[Open Data Program](http://www.seattle.gov/tech/initiatives/open-data/about-the-open-data-program)" in which it took the initiative to release all the data the City generates into its [Open Data Portal](https://data.seattle.gov/). The objective of this program is to [increase the quality of life of its residents]((http://www.seattle.gov/tech/initiatives/open-data/about-the-open-data-program)), so let's take advantage of that and see what we can find on collisions taking place in Seattle. 


The portal has a file called '[Collisions](https://data.seattle.gov/dataset/Collisions/nuam-5pkc)' with information containing all types of `collisions taking place from 2004 to the present day (September 15th, 2020)`. The data has 40 different attributes, and more information on this attributes can be found on the ['Attribute Information' (Metadata)](https://www.seattle.gov/Documents/Departments/SDOT/GIS/Collisions_OD.pdf) file.  

## Importing & Understanding our data <a name="Data_understanding"></a>


Let's import our data into a **Pandas DataFrame**, count the **attributes** (num of cols), and **observations** (num of rows) in our data. 

```python
#Import the crime data from the csv file and save it as a pandas dataframe
import pandas as pd 
Crime = pd.read_csv("..\Data\crime.csv", encoding='latin1')
Crime.shape
```

```python
Crime.columns.values
```

```python
Crime.dtypes
```

```python
Crime.head(5)
```

```python
Crime.dropna(subset=["OFFENSE_CODE_GROUP"], axis =0, inplace = True)
```

```python
import numpy as np
Crime.replace(np.nan,0, inplace=True)
```

Let's transform the 'OCCURRED_ON_DATE' to pandas date time format. 

```python
Crime['OCCURRED_ON_DATE'] = pd.to_datetime(Crime['OCCURRED_ON_DATE'])
```

Let's get the day of the week with the right format.

```python
#Get the day of the week when crime occurred, 0 means Moday and 6 means Sunday.
Crime['DAY_OF_WEEK']=Crime['OCCURRED_ON_DATE'].dt.dayofweek
```

Let's get the value counts for each type of offense

```python
Offenses = pd.DataFrame(Crime['OFFENSE_CODE_GROUP'].value_counts())
```

```python
Offenses.head(10)
```

```python
Offenses.tail(10)
```

```python
Crime.head()
```

#### Let's plot the behaviour of crime throughout the time. 

```python
Crime.groupby(['YEAR','MONTH']).size().rename('Crimes').reset_index().head(5)
```

```python
Crime.groupby(['YEAR','MONTH']).size().rename('Crimes').reset_index().tail(5)
```

Let's drop the observations from june 2015 and october 2019 since they don't seem to be complete.

```python
Crime.drop(Crime[(Crime['YEAR']==2015) & (Crime['MONTH']==6)].index, axis=0, inplace=True)
Crime.drop(Crime[(Crime['YEAR']==2019) & (Crime['MONTH']==10)].index, axis=0, inplace=True)
Monthly_data= Crime.groupby(['YEAR','MONTH']).size().rename('Crimes').reset_index()
Daily_data= Crime.groupby(['YEAR','DAY_OF_WEEK']).size().rename('Crimes').reset_index()
Hourly_data= Crime.groupby(['YEAR','HOUR']).size().rename('Crimes').reset_index()
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas #import FigureCanvas
from matplotlib.figure import Figure #import Figure artist
import seaborn as sns


fig = plt.figure() # create figure
fig.set_figwidth(14) # set width
fig.set_figheight(17) # set height
fig.suptitle(r"$\bf{ Figure 1:}$" " Seasonality of crimes by month, day of the week, and hour", y=0.09, fontsize=16)


ax0 = fig.add_subplot(3, 1, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(3, 1, 2) # add subplot 2 (1 row, 2 columns, second plot)
ax2 = fig.add_subplot(3, 1, 3) # add subplot 3 (1 row, 2 columns, third plot)

# Subplot 1: Box plot - Monthly
sns.boxplot(x=Monthly_data['MONTH'], y='Crimes', data=Monthly_data, showmeans=True, ax=ax0)
ax0.set_title('Boxplot of crimes by month(2015-2019)')
ax0.set_xlabel('Month')
ax0.set_ylabel('Number of crimes')

# Subplot 2: Box plot - Daily
sns.boxplot(x='DAY_OF_WEEK', y='Crimes', data=Daily_data, showmeans=True, ax=ax1)
ax1.set_title('Boxplot of crimes by day of the week(2015-2019)')
ax1.set_xlabel('Day of the week')
ax1.set_ylabel('Number of Crimes')

# Subplot 2: Box plot - Daily
sns.boxplot(x='HOUR', y='Crimes', data=Hourly_data, showmeans=True, ax=ax2)
ax2.set_title('Boxplot of crimes by hour of the day(2015-2019)')
ax2.set_xlabel('Hour of the day')
ax2.set_ylabel('Number of crimes')

plt.show()


#Title
fig.savefig('..\Figures\monthly_collisions.png')
```


We can see that there is indeed a seasonality of crimes, where at 5 p.m a maximum number of crimes is reached and the lowest point occurs at 4 a.m. On average, Fridays presents the highest number of crimes (up by ~10%), whilst Sunday presents the least (down by ~10%). From May to October crimes are relatively higher compared to the rest of the year. February is the month with the least number of crimes.

## Total number of crimes

```python
#We previously got the value counts for each type of crime and saved it as a data frame called "Offenses"
#Let's rename the value counts
Offenses.rename(columns={'OFFENSE_CODE_GROUP':'Total'},inplace=True)

#Let's add a descriptive name for the index column
Offenses.index.name='Offense_code'

#Finally let's sort values in descending order.
Offenses.sort_values(by='Total', ascending=False, axis=0, inplace=True)

#Let's get the sum of all the crimes.
print('The total number of offenses for the period is :', Offenses['Total'].sum(axis=0))
```

```python
Offenses.head(6)
```

```python
Offenses.tail(4)
```

We can see that from July, 2015 to September, 2019 a total number of 426, 839 offenses were recorded by the Boston Police Department (BPD). The most common incidents to which the BPD has responded are for a motor vehicle accident, larceny, medical assistance, to investigate a person, and drug violation.

```python
crimes_per_district.tail(3)
```

## Heat map of number of crimes per police district

We got the map of the [Police Districts in the City of Boston](https://data.boston.gov/dataset/police-districts) from 'ANALYZE BOSTON'. Let's import the geojson file with the geometry of each district and take a look to the Coordinates Reference System. 

```python
import geopandas as gpd
counties = gpd.read_file('..\Data\Police_Districts.geojson')
counties.crs
```

The coordinates are in the **World Geodetic System 1984** standard, which is used worldwide!

```python
counties.head(5)
```

The file has information on the name of the district, area, perimeter and coordinates of geometry points. Let's join the geometry with the number of crimes per district to create a heat map.

```python
import folium
```

```python
map_crimes = folium.Map([42.3239, -71.0842], zoom_start=12, tiles="OpenStreetMap")
crimes_per_district.reset_index(inplace=True)
map_crimes.choropleth(  geo_data=counties,
                    name='choropleth',
                    data=crimes_per_district,
                    columns=['DISTRICT', 'Total'],
                    key_on='feature.properties.ID',
                    fill_color='YlOrRd',
                    fill_opacity=0.7,
                    line_opacity=0.2,
                    legend_name='Crimes per district')
map_crimes
```

We can see that the area of Roxbury has the highest number of incidents. Let's create a heatmap that we can save as a png file.

```python
counties.set_index('DISTRICT',inplace=True)
crimes_per_district.set_index('DISTRICT', inplace=True)

counties_heatmap = counties.join(crimes_per_district)
```

```python
fig2 = plt.figure() # create figure
fig2.set_figwidth(8) # set width
fig2.set_figheight(8) # set height
fig2.suptitle(r"$\bf{ Figure 2:}$" " Heatmap of crimes in Boston", y=0.09, fontsize=14)


ax0 = fig2.add_subplot(1, 1, 1) # add subplot 1 (1 row, 2 columns, first plot)

# Subplot 1: Heat map - Crimes per district
counties_heatmap.plot(column='Total', cmap='OrRd', legend=True, ax=ax0)
ax0.set_title('Number of crimes per policial district in Boston')
ax0.set_xlabel('lon')
ax0.set_ylabel('lat')
counties.apply(lambda x: ax0.annotate(s=x.ID, xy=x.geometry.centroid.coords[0], ha='center'),axis=1);

plt.show()

#Title
fig2.savefig('..\Figures\heat_map_crimes.png')
```

So far we know that district B2 has the highest number of crimes from 2015 to 2019, but what type of crimes are most common in this area?
